## This notebook is having a quick prototype for predicting Hand written Data

```python
from __future__ import print_function
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD
from keras.utils import np_utils

np.random.seed(1671) # for reproducibility
```

```python
#Network Training 
NB_EPOCH = 200
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES = 10
OPTIMIZER = SGD() #SGD OPTIMIZER
VALIDATION_SPLIT = 0.2 # HOW MUCH TRAIN IS RESERVED FOR VALIDATION
```

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
#X_train is 60000 rows of 28x28 values --> reshaped in 60000 x 784

RESHAPED =  784
X_train = X_train.reshape(60000,RESHAPED)
X_test = X_test.reshape(10000,RESHAPED)
```

```python
X_train = X_train.astype("float32")
X_test  =X_test.astype("float32")
```

```python
#Normalize the data set
X_train = X_train/255
X_test =  X_test/255
```

```python
#Lable conversion from categorical to ONE HOT Encoded 
import numpy as np
Y_train = np_utils.to_categorical(y_train,NB_CLASSES)
Y_test = np_utils.to_categorical(y_test,NB_CLASSES)
```

```python
# 10 outputs
# final stage is softmax
#Making model

model = Sequential()
model.add(Dense(NB_CLASSES,activation="relu",input_shape=(RESHAPED,)))
model.add(Activation("softmax"))
model.summary()

model.compile(loss = "categorical_crossentropy",optimizer = SGD() , metrics = ["accuracy"] )
model.fit(X_train,Y_train, batch_size = BATCH_SIZE ,epochs = NB_EPOCH,verbose = VERBOSE , validation_split=VALIDATION_SPLIT)
```

```python
score = model.evaluate(X_test, Y_test, verbose=VERBOSE)
print("Test score:", score[0])
print('Test accuracy:', score[1])
```
